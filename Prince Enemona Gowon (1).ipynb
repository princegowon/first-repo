```python
# Identify redundant columns
redundant_cols = ["Case Number"]  
# Drop redundant columns
df_cleaned = df.drop(columns=redundant_cols)
```

```python
df_cleaned
```

```python
# Extract month, day,"Date" column
df_cleaned["Month"] = pd.to_datetime(df_cleaned["Date"], format="%m/%d/%Y %I:%M %p").dt.month_name()  #Extracts month
df_cleaned["Day"] = pd.to_datetime(df_cleaned["Date"], format="%m/%d/%Y %I:%M %p").dt.day_name()   #extracts day
```

```python
# creating a column for seasons

def create_seasons_column(date_column):
    """
    Creates a new column representing seasons based on the months in a datetime column.

    Parameters:
    - date_column: Pandas Series, representing a datetime feature

    Returns:
    - Pandas Series containing the corresponding season names
    """
    # coverting  input to  datetime format
    date_column = pd.to_datetime(date_column, format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

    # Define a mapping of the month to season
    month_to_season = {
        1: 'Winter', 2: 'Winter', 3: 'Spring',
        4: 'Spring', 5: 'Spring', 6: 'Summer',
        7: 'Summer', 8: 'Summer', 9: 'Fall',
        10: 'Fall', 11: 'Fall', 12: 'Winter'
    }

    # Extract the month and map to season
    seasons_column = date_column.dt.month.map(month_to_season)

    return seasons_column
```

```python

df_cleaned['Season'] = create_seasons_column(df_cleaned['Date'])
```

```python
# Group by Months, Days, Seasons and count crimes
crime_counts_month = df_cleaned.groupby("Month")["Primary Type"].count()
crime_counts_day = df_cleaned.groupby("Day")["Primary Type"].count()
crime_counts_season = df_cleaned.groupby("Season")["Primary Type"].count()

# Find months, days, seasons with highest and lowest crime
highest_crime_month = crime_counts_month.idxmax()
lowest_crime_month = crime_counts_month.idxmin()
highest_crime_day = crime_counts_day.idxmax()
lowest_crime_day = crime_counts_day.idxmin()
highest_crime_season = crime_counts_season.idxmax()
lowest_crime_season = crime_counts_season.idxmin()

# Find highest and lowest crimes by frequency
highest_crimes = df_cleaned["Primary Type"].value_counts().head(5).index.tolist()
lowest_crimes = df_cleaned["Primary Type"].value_counts().tail(5).index.tolist()

# Print results
print(f"Month with highest crime: {highest_crime_month}")
print(f"Month with lowest crime: {lowest_crime_month}")
print(f"Day with highest crime: {highest_crime_day}")
print(f"Day with lowest crime: {lowest_crime_day}")
print(f"Season with highest crime: {highest_crime_season}")
print(f"Season with lowest crime: {lowest_crime_season}")
print(f"Highest occurrence crimes: {highest_crimes}")
print(f"Lowest occurrence crimes: {lowest_crimes}")
```

```python
# Group by location description and count crimes
crime_counts_location = df_cleaned.groupby("Location Description")["Primary Type"].count()

# Find location with highest crime
highest_crime_location = crime_counts_location.idxmax()

# Print result
print(f"Location with highest crime: {highest_crime_location}")
```

```python
# Group by year and primary type, then count crimes
crime_counts_year = (
    df_cleaned.groupby(["Year", "Primary Type"])["ID"]
    .count()
    .reset_index()
    .rename(columns={"ID": "Count"})
)

# Pivot table to show crime type evolution across years
pivot_table = crime_counts_year.pivot_table(
    index="Year", columns="Primary Type", values="Count", aggfunc="sum"
)
```
